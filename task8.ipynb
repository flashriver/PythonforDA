图形绘制.
图形对象被称为块——patch，例如Rectangle，Circle等
matplotlib.pyplot有一些常见的图形,完整集合位于matplotlib.patches

```python
fig=plt.figure()
```

```python
ax=fig.add_subplot(1,1,1,)
```

```python
rect=plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3)
```

```python
circ=plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)
```

```python
pgon=plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5)
```

上边只是创建了对象，并没有实际进行绘制。要绘制图形，需要使用subplot对象的add_patch方法，并传入需要绘制的图形对象的名称

```python
ax.add_patch(rect)
```

```python
ax.add_patch(circ)
```

```python
ax.add_patch(pgon)
```

利用上述方法可以绘制多种图形，然后可以组装成各种好看的图表对象。就像是用基础的乐高积木搭建各种物体一样。

$\lim \limits{(n\rightarrow \infty)} \frac{1}{x-\frac{1}{n}}$

```python
fig1=plt.figure()
```

```python
ax=fig1.add_subplot(1,1,1,)
```

```python
ax.add_patch(plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3))
```

```python
ax.add_patch(plt.Rectangle((0.1,0.3),0.4,0.15,color='b',alpha=0.3))
```

```python
ax.add_patch(plt.Rectangle((0.1,0.6),0.2,0.35,color='r',alpha=0.3))
```

```python
ax.set_xticks([1,3,5,9])
#修改横坐标刻度，图形随之改变了
#前边搞错了吗？
```

### 分面网格（facet grid）和类型数据

如果数据有额外的分组维度，有多个分类变量的数据可视化的一种方法是，使用小面网格。

seaborn的内置函数factorplot可以简化制作多种分面图。

```python
fig=plt.figure();
ax=fig.add_subplot(1,1,1);
```

```python
sns.factorplot(x='day',y='tip_pct',hue='time',col='smoker',kind='bar',data=tips[tips.tip_pct<1])
#factorplot方法会创建新的fig对象并在上边画图
#x,y参数指定了横、纵坐标所对应的列
#hue参数指定了按照time列的值进行分项统计
```

```python
sns.factorplot(x='day',y='tip_pct',row='time',col='smoker',kind='bar',data=tips[tips.tip_pct<1])
#在上述语句基础上，修改,hue='time'为row='time'
#hue='time'参数将time不一样的列并列显示
#row='time'则修改为根据time列的情况，分行显示，每个subplot里只显示一个时间的统计信息
```

```python
sns.factorplot(x='tip_pct',y='day',kind='box',data=tips[tips.tip_pct<0.5])
#盒图——类似于SPSS的箱图？
```
